```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
1468
3667
start=3752
end=3752


ncolumns=['Lote','Sexo','Cantidad','Peso Total','Peso Promedio','Procedencia','Observaciones','Hora Entrada','Precio Base', 'Precio Final','Fecha','Ciudad','Tipo','Martillo','N°Feria']
Gan=pd.DataFrame(columns=ncolumns)
url = "https://www.subagan.com/site/index.php/lot-reports/view/{url_number}"
for url_number in range(start, end + 1):
    url1 = url.format(url_number=url_number)
   
    r = requests.get(url1)

    soup = BeautifulSoup(r.content, 'html.parser')
    tables = soup.find_all('table',class_ = "table")
    table_data = []
    tables_dict = {}
  # Iterate through each table and extract data
    for idx, table in enumerate(tables):
        table_rows=[]
        # Iterate through rows and columns of the table
        for row in table.find_all('tr'):
            # Extract data from each cell in the row
            cells = row.find_all('td')  # Change 'td' to 'th' if you want to include header cells
            row_data=[cell.text.strip() for cell in cells]
            table_rows.append(row_data)
      

        table_data.append(table_rows)
    
    dfs = [pd.DataFrame(data) for data in table_data]
    df1=dfs[0]  
    df2=dfs[1]
    Fecha=df1.iloc[1,0].strip('Fecha')
    Ciudad=df1.iloc[0,2].strip('Ciudad')
    Tipo=df1.iloc[0,1].strip('Tipo')
    Martillo=df1.iloc[1,1].strip('Martillo')
    NFeria=df1.iloc[0,0].strip('Feria')
# # table_key = f'table{idx + 1}'
# # tables_dict[table_key] = pd.DataFrame(table_rows)

    df2['Fecha']=Fecha
    df2['Ciudad']=Ciudad
    df2['Tipo']= Tipo
    df2['Martillo']=Martillo
    df2['N°Feria']=NFeria

# with iloc we can separate the two tables, and extact the date and other info to create a new clean format table.
    df2.dropna(inplace=True)
    df2.rename(columns= {0:'Lote',1:'Sexo',2:'Cantidad',3:'Peso Total',4:'Peso Promedio',5:'Procedencia',6:'Observaciones',7:'Hora Entrada',8:'Precio Base', 9:'Precio Final',10:'Fecha',11:'Ciudad',12:'Tipo',13:'Martillo',14:'N°Feria'}, inplace=True)
    Gan=pd.concat([Gan,df2],ignore_index=True)
    
    


Gan.replace(r'^\s*$', np.nan, regex=True, inplace=True)
Gan.replace(0, np.nan, regex=True, inplace=True)
Gan.dropna(inplace=True, subset=['Precio Final', 'Precio Base'])
Gan['Precio Base']=Gan['Precio Base'].astype(str)
Gan['Precio Final']=Gan['Precio Final'].astype(str)
Gan['Peso Promedio']=Gan['Peso Promedio'].astype(int)
Gan['Cantidad']=Gan['Cantidad'].astype(int)
Gan['Precio Final']=Gan['Precio Final'].str.replace('.','')
Gan['Precio Base']=Gan['Precio Base'].str.replace('.','')
Gan['Precio Final']=Gan['Precio Final'].str.replace(',','')
Gan['Precio Base']=Gan['Precio Base'].str.replace(',','')
Gan['Precio Base']=Gan['Precio Base'].astype(int)
Gan['Precio Final']=Gan['Precio Final'].astype(int)

miss= ((Gan['Precio Base']>0)&(Gan['Precio Final']>0)&(Gan['Precio Final']<20000)&(Gan['Precio Base']<20000))
Gan=Gan.loc[miss]


df=Gan
df
```

```python
LabelS=df['Sexo'].unique().astype(str)
LabelS
X=df[['Sexo','Peso Promedio','Precio Final','Cantidad']]
X
result_table = pd.DataFrame()

# Iterate over each asset label
for i in LabelS:
    # Get the prices for the current asset
    Peso_Promedio= X[X['Sexo']==i] ['Peso Promedio']
    Precio_Final= X[X['Sexo']==i]['Precio Final']
    cantidad=X[X['Sexo']==i]['Cantidad']
    # Calculate the range and mean
    Peso_range = (min(Peso_Promedio), max(Peso_Promedio) + 1)
    Peso_mean = Peso_Promedio.mean()
    Precio_range=(min(Precio_Final), max(Precio_Final) + 1)
    Precio_mean = Precio_Final.mean()
    Cantidad= sum(cantidad)
    # Add a new row to the result table
    result_table = result_table.append({
        'Sexo': i,
        'Rango Peso ': Peso_range,
        'Media Peso ': round(Peso_mean),
        'Rango Precio':Precio_range,
        'Media Precio':round(Precio_mean),
        'Cantidad':Cantidad
    }, ignore_index=True)

    
result_table
```
